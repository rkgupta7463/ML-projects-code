```python
final_df['Rating Description'].isnull().sum()
```

```python
final_df['Reviews'].isnull().sum()
```

```python
final_df['Star Rating'].mean()
```

```python
final_df['Star Rating']=final_df['Star Rating'].fillna(4.010)
```

```python
final_df['Star Rating'].isnull().sum()
```

```python
final_df['Nearest Landmark'].mode()
```

```python
final_df['Nearest Landmark']=final_df['Nearest Landmark'].fillna('T2 - Chhatrapati Shivaji International Airport')
```

```python
final_df['Nearest Landmark'].isnull().sum()
```

```python
final_df['Distance to Landmark'] = final_df['Distance to Landmark'].str.replace('km', '').str.replace('m', '')
```

```python
final_df['Distance to Landmark'].isnull().sum()
```

```python
final_df['Distance to Landmark'].mode()
```

```python
final_df['Distance to Landmark']=final_df['Distance to Landmark'].fillna('3.2')
```

```python
final_df['Distance to Landmark'].isnull().sum()
```

```python
final_df['Distance to Landmark']=final_df['Distance to Landmark'].astype(float)
```

```python
final_df['Tax'].isnull().sum()
```

```python
final_df['Tax'].mode()
```

```python
final_df['Tax']=final_df['Tax'].fillna('770')
```

```python
final_df['Tax'].isnull().sum()
```

```python
final_df['Price']=final_df['Price'].str.replace(',','').astype(int)
```

#### 3. Data Splitting

```python
X=final_df.drop(columns=['Price'])
y=final_df['Price']
```

```python
from sklearn.model_selection import train_test_split
##preproccesing modules 
from sklearn.preprocessing import OneHotEncoder,StandardScaler
##columnTransform 
from sklearn.compose import ColumnTransformer
##pipelines
from sklearn.pipeline import Pipeline
##importing the required models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor,XGBRFRegressor
##importing the requires evaluation matrics
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,mean_absolute_percentage_error
import math
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

#### 4. Feature Engineering

```python
onh=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
stc=StandardScaler()
```

```python
##creating the columnstransformer
tf1=ColumnTransformer(transformers=[
    ('OneHot Encoder',onh, [0,2,5,6])
],remainder='passthrough')
```

```python
tf2=ColumnTransformer(transformers=[
    ('StandardScaler',stc,slice(0,None))
],remainder='passthrough')
```

### **M-1**

```python
lr=LinearRegression()
pipe_lr=Pipeline([
    ('tf1',tf1),
    ('tf2',tf2),
    ('lr',lr)
])
```

```python
pipe_lr.fit(X_train,y_train)
```

```python
y_pred_lr=pipe_lr.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred_lr)
mae = mean_absolute_error(y_test, y_pred_lr)
r2 = r2_score(y_test, y_pred_lr)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {math.sqrt(mse)}")
print(f"R-squared (R2) Score: {r2}")
```

```python
plt.scatter(y_test, y_pred_lr)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Scatter Plot of Actual vs. Predicted Values")
plt.show()
```

```python
residuals = y_test - y_pred_lr
plt.scatter(y_pred_lr, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
plt.plot(y_test, label="Actual Values")
plt.scatter(range(len(y_test)), y_pred_lr, label="Predicted Values", marker='o', color='r')
plt.xlabel("Data Points")
plt.ylabel("Values")
plt.title("Line Plot of Actual vs. Predicted Values")
plt.legend()
plt.show()
```

```python
sns.histplot(y_test, label="Actual", kde=True)
sns.histplot(y_pred_lr, label="Predicted", kde=True)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Distribution of Actual vs. Predicted Values")
plt.legend()
plt.show()
```
